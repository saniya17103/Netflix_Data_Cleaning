## Netflix Movies and Tv Shows Data Cleaning

```python
## Import libraries
import pandas as pd
import numpy as np
```

```python
## Load dataset
df = pd.read_csv("netflix_titles.csv.zip")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
## check missing values
df.isnull().sum()
```

```python
## Handle missing values
df[df['director'].isnull()].head()
```

```python
## Fill missing values
df['director'] = df['director'].fillna("Unknown")

df['cast'] = df['cast'].fillna("Unknown")

df['country'] = df['country'].fillna("Unknown")

df['duration'] = df['duration'].fillna("Unknown")
```

```python
df['rating'].mode()
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
## Remove missing dates
df = df.dropna(subset=['date_added'])
```

```python
## Now check againg
df.isnull().sum()
```

```python
## Check Duplicate Records
df.duplicated().sum()
```

```python
## check inconsistent values
df.nunique()
```

```python
## Standardize text values
df['type'].unique()
```

```python
df['country'].unique()[:20]
```

```python
# Remove extra spaces in country column
df['country'] = df['country'].str.strip()

# Standardize capitalization
df['country'] = df['country'].str.title()
```

```python
## check
df['country'].unique()[:20]
```

```python
df['country'].str.contains('  ').sum()
```

```python
## find wrong rating values
df['rating'].unique()
```

```python
## lets move in duration column
df.loc[df['rating'].isin(['74 min','84 min','66 min']), 'duration'] = df['rating']

df.loc[df['rating'].isin(['74 min','84 min','66 min']), 'rating'] = np.nan
```

```python
df.isnull().sum()
```

```python
## fill missing values in rating
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
## check
df['rating'].unique()
```

```python
## check data types
df.dtypes
```

```python
df.info()
```

```python

df['date_added'].head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['date_added'].isnull().sum()
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['date_added'].dtype
```

```python
df.info()
```

```python
## check column names
df.columns
```

```python
## Convert columns
df.columns = df.columns.str.lower().str.replace(" ","_")
```

```python
df.columns
```

```python
## convert date_added to consistent format
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```
